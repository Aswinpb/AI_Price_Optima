```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/processed/sales_cleaned.csv")

df.head()
```

Revenue Column

```python
df["Revenue"] = df["Units Sold"] * df["Price"]
```

Price vs Quantity Sold

```python
# Create price bins
df["Price_Bin"] = pd.cut(df["Price"], bins=10)

# Calculate average units sold per price bin
price_bin_trend = df.groupby("Price_Bin")["Units Sold"].mean()

plt.figure(figsize=(8,5))
price_bin_trend.sort_index().plot(kind="bar")
plt.title("Average Units Sold by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Average Units Sold")
plt.xticks(rotation=45)
plt.show()
```

This chart groups prices into ranges and shows the average demand for each range.
The demand remains relatively stable across price ranges, indicating low price
elasticity in the dataset. This suggests that other factors such as promotions,
inventory levels, or competitor pricing may influence demand more than price.

Sales Over Time


```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
sales_time = df.groupby("Date")["Units Sold"].sum()

plt.figure(figsize=(10,5))
sales_time.plot()
plt.title("Sales Over Time")
plt.ylabel("Units Sold")
plt.show()
```

This plot shows how total units sold change over time.
Fluctuations indicate changing demand patterns across different periods.
Understanding these trends helps identify seasonal demand.


Revenue Over Time

```python
revenue_time = df.groupby("Date")["Revenue"].sum()

plt.figure(figsize=(10,5))
revenue_time.plot()
plt.title("Revenue Over Time")
plt.ylabel("Revenue")
plt.show()
```

Revenue varies across time periods due to changes in price and sales volume.
High peaks indicate periods where both demand and price contributed to higher revenue.

Inventory vs Sales

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="Inventory Level", y="Units Sold", data=df)
plt.title("Inventory Level vs Units Sold")
plt.show()
```

This visualization shows the relationship between inventory levels and sales.
Higher inventory availability often corresponds with higher units sold,
indicating inventory constraints may limit sales.

Product-wise Sales Distribution

This chart shows the top selling products by total units sold.
A few products dominate sales, suggesting opportunities for product-based
pricing strategies.

Price Distribution

Load Processed Data

```python
import pandas as pd

df = pd.read_csv("../data/processed/sales_cleaned.csv")

df.head()
```

Calculate Original Revenue (Static Pricing)

Revenue = Price × Units Sold

```python
df["Original_Revenue"] = df["Price"] * df["Units Sold"]

original_revenue = df["Original_Revenue"].sum()

print("Original Total Revenue:", original_revenue)
```

 Dynamic Pricing Rules

Created a new price column Adjusted_Price

```python
df["Adjusted_Price"] = df["Price"]
```

Inventory-Based Pricing Rules

```python
# Low inventory → increase price
df.loc[df["Inventory Level"] < 50, "Adjusted_Price"] *= 1.15

# High inventory → decrease price
df.loc[df["Inventory Level"] > 300, "Adjusted_Price"] *= 0.90
```

Time-Based Pricing Rules

date column

```python
df["Date"] = pd.to_datetime(df["Date"])
```

Day of week

```python
df["DayOfWeek"] = df["Date"].dt.dayofweek
```

Weekend → increase price 10%

```python
df.loc[df["DayOfWeek"] >= 5, "Adjusted_Price"] *= 1.10
```

Calculate New Revenue (Dynamic Pricing)

```python
df["New_Revenue"] = df["Adjusted_Price"] * df["Units Sold"]

new_revenue = df["New_Revenue"].sum()

print("New Total Revenue:", new_revenue)
```

Calculate Revenue Lift

Revenue Lift =
(New Revenue - Original Revenue) / Original Revenue × 100

```python
revenue_lift = ((new_revenue - original_revenue) / original_revenue) * 100

print("Revenue Lift (%):", revenue_lift)
```

Show Pricing Comparison

```python
df[["Price","Adjusted_Price","Units Sold","Original_Revenue","New_Revenue"]].head()
```

Visualization

```python
import matplotlib.pyplot as plt

labels = ["Static Pricing", "Dynamic Pricing"]
values = [original_revenue, new_revenue]

plt.bar(labels, values)
plt.title("Revenue Comparison")
plt.ylabel("Total Revenue")
plt.show()
```

Pricing Logic Explanation

Inventory-based pricing increases prices when stock is low and decreases prices when inventory is high.

Time-based pricing increases prices during weekends to capture higher demand periods.

The baseline pricing engine simulates simple dynamic pricing rules before applying machine learning models.

PRICE CHANGES

```python
price_changes = (df["Price"] != df["Adjusted_Price"]).sum()

print("Number of prices adjusted:", price_changes)
```
