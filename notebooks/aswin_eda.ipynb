```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns

df = pd.read_csv("D:/infosys/AI_Price_Optima/data/retail_store_inventory.csv")
df.head()
```

Revenue Column

```python
df["Revenue"] = df["Price"] * df["Units Sold"]
df.head()
```

Check Correlation

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

Price vs Units Sold


```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="Price", y="Units Sold", data=df, alpha=0.3)
plt.title("Price vs Units Sold")
plt.show()
```

Discount vs Units Sold


```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="Discount", y="Units Sold", data=df, alpha=0.3)
plt.title("Discount vs Units Sold")
plt.show()
```
